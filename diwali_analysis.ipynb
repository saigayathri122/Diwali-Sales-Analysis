```python
#dropping blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
#checking for null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount']
```

```python
#Rename columns for clear understanding.
df.rename(columns = {'Shaadi':'Marital_Status'})
```

```python
#checking the statistical values for the parameters.
df[['Age','Orders','Amount']].describe()
```

```python
#Generating a box plot for gender.
ax = sns.countplot(x = 'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
#concluded that female are tuned to shop more.
```

```python
#Finding the gender wise amount spent for shopping.
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bars in ax.containers:
    ax.bar_label(bars)
#concluded that women are spending more for shopping.
```

From the above graphs we can see that most of the buyers are females and the purchasing power of females is still greater than males

**AGE**

```python
#Finding the age group who shops more.
ax = sns.countplot(x= df['Age Group'],data=df,hue=df['Gender'])
for bars in ax.containers:
    ax.bar_label(bars)
#Concluded that the age group 26-35 and mainly females are mostly shopping.
```

```python
#Segregating the data as age group and salary for that age group.
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot(x='Age Group',data=sales_age,y='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
#Concluded that the age group between 26-35 are shopping more than others.
```

From the above graphs we can see that most of the buyers are from the age group 26-35 female

```python
#Finding the state that orders more(sorted by number of orders).
state_sales = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.barplot(data=state_sales,x='State',y='Orders')
#Concluded that the Uttar Pradesh state dominate the others.
```

```python
#Finding the state that orders more(sorted by sum of amount).
state_sales = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.barplot(data=state_sales,x='State',y='Amount')
#Concluded that the Uttar Pradesh state dominate the others.
```

from the above graphs we can infer that the more number of orders as well as the purchasing power is greater for Uttar Pradesh state followed by Maharastra and Karnataka

**Marital Status**

```python
#finding the marital status of the people who are ordering more.
ax = sns.countplot(x=df['Marital_Status'],data=df)
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
#the unmarried are ordering more.
```

```python
#who are ordering more when compared with marital status??
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
#unmarried women are ordering more.
```

*From the above graphs we can see that most of the buyers are married women and they also have high purchasing power*

**OCCUPATION**

```python
#People in the IT Sector are ordering more.
ax = sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_occ,x='Occupation',y='Amount')
```

From the above graph we can conclude that most of the buyers are from IT,HealthCare,Aviation

**Product Category**

```python
ax = sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
#Clothing and Apparel stands at the first place.
```

```python
sales_pro = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax = sns.barplot(data=sales_pro,x='Product_Category',y='Amount')
```

from the above grpahs we can infer that the most sold product category is food,Clothing and apparel,electronics and Gadgets according to the amount

**MOST SOLD PRODUCT**

```python
most_sold = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=most_sold,x='Product_ID',y='Orders')
```

**FINAL CONCLUSION**

**Married Women of age 26-35 yrs from UP,Maharastra,Karnataka working in IT,Healthcare and Aviation are more likely to buy products from the food,clothing and electronics Category**
